## Principal Component Analysis (or PCA)

Principal Component Analysis (or PCA) uses linear algebra to transform the dataset into a compressed form.It is not a full machine learning algorithm but instead an unsupervised learning algorithm. 

PCA is a technique for feature extraction — so it combines our input variables in a specific way, then we can drop the “least important” variables while still retaining the most valuable parts of all of the variables! As an added benefit, each of the “new” variables after PCA are all independent of one another. 


1. As the number of dimensions increases it is a curse and accuracy got decreses. 
2. By reducing the dimension of  the feature space, we have fewer relationships between variables to consider and which is less likely to overfit your model.
3. There are many ways to achieve dimensionality reduction, but most of these techniques fall into one of two classes:

- Feature Elimination
- Feature Extraction

4. First principal component is a linear combination of original predictor variables which captures the maximum variance in the data set. It determines the direction of highest variability in the data. Larger the variability captured in first component, larger the information captured by component. No other component can have variability higher than first principal component.

The first principal component results in a line which is closest to the data i.e. it minimizes the sum of squared distance between a data point and the line.

Similarly, we can compute the second principal component also.PCA finds the best fitting line by maximizing the sum of the squared distances from the projected points to the origin. - using pythagoras theorem- a2 + b2 = c2.

Second principal component (Z²) is also a linear combination of original predictors which captures the remaining variance in the data set and is uncorrelated with Z¹. In other words, the correlation between first and second component should is zero. 




PCA is a method that brings together:

A measure of how each variable is associated with one another. (Covariance matrix.)
The directions in which our data are dispersed. (Eigenvectors.)
The relative importance of these different directions. (Eigenvalues.)
PCA combines our predictors and allows us to drop the eigenvectors that are relatively unimportant.


### Loading the dataset

```python
import numpy as np

from pandas import read_csv
from sklearn.decomposition import PCA
# load data
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = read_csv(url, names=names)
```

```python
dataframe
```

```python
X = dataframe.drop(['class'], axis =1)
```

```python
X.shape
```

```python
y = dataframe['class']
```

```python
y.shape
```

### Feature extraction using PCA

Here we will apply PCA to select revelant features and fit the function. 

```python
# feature extraction
pca = PCA(n_components=3)
fit = pca.fit(X)
```

```python
# summarize components
print("Explained Variance: %s" % fit.explained_variance_ratio_)
#print(fit.components_)
components = pca.fit_transform(X)
```

This shows that first principal component explains 88.8% variance. Second component explains 6.1% variance. Third component explains 2.5% variance and so on. So, how do we decide how many components should we select for modeling stage. The answer is in the plot. we will plot the components and see the relationship.

### Plotting the principal components

```python
import plotly.express as px

labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(fit.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(3),
    color=dataframe['class']
)
fig.update_traces(diagonal_visible=False)
fig.show()
```

The idea behind pca is to construct some principal components which satisfactorily explains most of the variability in the data, as well as relationship with the response variable. We can use PCA in any machine learning algorithm for dimensionality reduction.
